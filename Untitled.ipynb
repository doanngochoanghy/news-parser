```python
import glob
import pandas as pd
import codecs
dfs = []
for i in glob.glob('./data/vnexpress/*.json'):
    df = pd.read_json(codecs.open(i,'r','utf-8'))
    print(i)
    dfs.append(df)
```

```python
data = pd.concat(dfs,ignore_index=True)
data['news_id'] = data['news_id'].astype(str)
```

```python
data_news = data
```

```python
# data_news = data_news[(data_news['title_entities'].str.len() > 0) & (data_news['abstract_entities'].str.len() > 0)]
data_news = data_news[~data_news.comments.isnull()]
data_news = data_news[(data_news.timestamp > '2021-01-01')]
data_comments = data_news.explode('comments')
data_comments = data_comments[~data_comments.comments.isnull()]
```

```python
data.count()
data_news.count()
```

```python
data_comments
```

```python
data_comments['user_id'] = data_comments['comments'].apply(lambda x:x.get('user_id'))
```

```python
data_comments.groupby(['user_id']).count()
```

```python
data_users = []
data_comments_by_user = data_comments.groupby(['user_id'])
count = 0
for user_id, data in data_comments_by_user:
    news_ids = []
    if user_id == 'None':
        continue
    if len(data) < 5 or len(data) > 200:
        continue
    history_news = data.sample(frac=0.8)
    clicked_news = data[~data.news_id.isin(history_news.news_id)]
    history_news_data = (' '.join(history_news.news_id))
    for i, news in clicked_news.iterrows():
        news_date  = news.timestamp.date()
        impression_news = data_news[data_news.timestamp.dt.date == news_date].sample(frac=0.3)
        impression_news_data = ' '.join([news_id + '-0' for news_id in impression_news.news_id]) + ' '+ news.news_id+'-1'
        users = {'user_id': user_id, 'history_news_data': history_news_data,"impression_news_data": impression_news_data, 'timestamp':news.timestamp}
        data_users.append(users)
    count += 1 
    if count % 100 ==0:
        print(count, len(data_users))
```

```python
data_users = pd.DataFrame(data_users)
```

```python
from sklearn.model_selection import train_test_split
train_users, test_users = train_test_split(data_users, test_size=0.3)
```
